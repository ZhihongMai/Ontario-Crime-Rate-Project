# Cleaning the Data

````python
# Standardizing Geography Columns Across Datasets

crime_demogrp['Geography'] = crime_demogrp['Geography'].str.replace(r',\sOntario.*', ', Ontario', regex=True)
````

````python
# Standardizing Link Geography Columns Across Datasets

criminal['Link Geography'] = criminal['Link Geography'].str.replace(r',\sOntario.*', ', Ontario', regex=True)
````

````python
# Checking if Both Columns have the Same Unique Values

criminal['Link Geography'].nunique()
````

````python
# Checking if Both Columns have the Same Unique Values

crime_demogrp['Geography'].nunique()
````

````python
# Creating a Province Column from Link Geography in criminal Dataset

criminal['province'] = criminal['Link Geography'].str.split(', ').str[-1].str.strip()
````

````python
# Checking if the Province Column Only Has 1 Unique Value

criminal['province'].nunique()
````

````python
# Standardizing OFFENCE GROUP Values

criminal['OFFENCE GROUP'] = criminal['OFFENCE GROUP'].replace({'Admin of Justice': 'Administration of Justice'})
````

````python
# Filling NA with Column Average for Crime Demographic

crime_demogrp['Census families in private households'] = crime_demogrp['Census families in private households'].fillna(
    crime_demogrp['Census families in private households'].mean())
````

````python
# Replacing Missing Values with 0 for Criminal Dataset

criminal.fillna(0,inplace=True)
````

````python
# Converting Trial Rate to Numeric Format

criminal['Trial Rate'] = criminal['Trial Rate'].str.replace('%', '').astype(float)
````

````python
# Merging Crime Demogrp and Criminal Datasets

complete_criminal_df = pd.merge(crime_demogrp, criminal, left_on = 'Geography', right_on = 'Link Geography', how = 'inner')
````

````python
# Categorizing Family, Population, Economic, and Education Factor Columns into Low, Average, and High

def categorize_column(df, column_name):
    for column in column_name: 
        q1=df[column_name].quantile(0.25)
        q3=df[column_name].quantile(0.75)
        
        def categorize(value):
            if value < q1:
                return 'Low'
            elif value <= q3:
                return 'Average'
            else:
                return 'High'
        return df[column_name].apply(categorize)
````

````python
# Categorizing Lone-Parent Families Percentage into Groups

complete_criminal_df['Lone-parent Families %'] = categorize_column(
    complete_criminal_df, 'Lone-parent families, proportion of census families (percent)')
````

````python
# Categorizing Female Lone-Parent Families Percentage into Groups

complete_criminal_df['female lone-parent families %'] = categorize_column(
    complete_criminal_df, 'Female lone-parent families, proportion of census families (percent)')
````

````python
# Categorizing Male Lone-Parent Families Percentage into Groups

complete_criminal_df['Male lone-parent families %'] = categorize_column(
    complete_criminal_df, 'Male lone-parent families, proportion of census families (percent)')
````

````python
# Categorizing Avg Total Income into Groups

complete_criminal_df['Avg total income in 2010'] = categorize_column(
    complete_criminal_df, 'Average total income in 2010 of population 15 years and over (canadian dollars)')
````

````python
# Categorizing Low Income before Tax Families Percentage into Groups

complete_criminal_df['low income before tax in 2010'] = categorize_column(
    complete_criminal_df, 'Prevalence of low income before tax in 2010 for economic families for income status (percent)')
````

````python
# Categorizing 17 Years Live in Low Income Percentage into Groups

complete_criminal_df['17 yrs live in low income'] = categorize_column(
    complete_criminal_df, 'Prevalence of persons aged 17 years and under living in low income economic families before tax in 2010 (percent)')
````

````python
# Categorizing Unemployment Rate Percentage into Groups

complete_criminal_df['Unemployment rate'] = categorize_column(
    complete_criminal_df, 'Unemployment rate 15 years and over (percent)')
````

````python
# Categorizing Immigrant Population Percentage into Groups

complete_criminal_df['Immigrant population %'] = categorize_column(
    complete_criminal_df, 'Immigrant population, proportion of total population (percent)')
````

````python
# Categorizing Immigrant Population Arrived 2001-2011 Percentage into Groups

complete_criminal_df['Immigrant population arrived (2001-2011) %'] = categorize_column(
    complete_criminal_df, 'Immigrants who arrived between 2001 and 2011, proportion of total population (percent)')
````

````python
# Categorizing Visible Minority Population Percentage into Groups

complete_criminal_df['Visible minority population %'] = categorize_column(
    complete_criminal_df, 'Visible minority population, proportion of total population (percent)')
````

````python
# Categorizing High School Graduates Percentage into Groups

complete_criminal_df['High school graduates %'] = categorize_column(
    complete_criminal_df, 'High school graduates aged 25 to 29, proportion of population aged 25 to 29 (percent)')
````

````python
# Categorizing Post-Secondary Graduates Percentage into Groups

complete_criminal_df['Post-secondary graduates %'] = categorize_column(
    complete_criminal_df, 'Post-secondary graduates aged 25 to 54, proportion of population aged 25 to 54 (percent)')
````

````python
# Function for Correlation Matrix and Rename Columns and Indexs
def correlation_rename(df, columns, rename_dict):
    corr = df[columns].corr()
    corr = corr.rename(columns = rename_dict, index = rename_dict)
    return corr
````

````python
# Create a Bins Function to Cut Dataframe

def bins_cut(df, column, bins, new_column):
    df[new_column] = pd.cut(df[column], bins=bins)
````

````python
# Immigrant population bins
bins_cut(complete_criminal_df, 'Immigrant population, proportion of total population (percent)', 3, 'Immigrantion')
immigrantion_df = complete_criminal_df.groupby('Immigrantion', observed=False)['Cases Received'].mean().reset_index()

# Immigrant population 2001-2011 bins
bins_cut(complete_criminal_df, 'Immigrants who arrived between 2001 and 2011, proportion of total population (percent)', 4, 'Immigrantion 2001-2011')
immigrantion_2001_df = complete_criminal_df.groupby('Immigrantion 2001-2011', observed=False)['Cases Received'].mean().reset_index()

# Minority_bins
bins_cut(complete_criminal_df, 'Visible minority population, proportion of total population (percent)', 3, 'Visible minority population')
minority_df = complete_criminal_df.groupby('Visible minority population', observed=False)['Cases Received'].mean().reset_index()
````

````python
# Create a subplot to See Average Cases Received Across Population Factors

fig, (ax1,ax2,ax3) = plt.subplots(nrows = 3, ncols =1, figsize = (10,10))

sns.barplot(x='Immigrantion', y='Cases Received', data= immigrantion_df, hue='Cases Received', palette='coolwarm', legend=False, ax=ax1)
ax1.set_title('Average Cases Received by Immigrantion of Total Population')
ax1.set_xlabel('Immigrantion of Total Population (%)')
ax1.set_ylabel('Average Cases Received')

sns.barplot(x='Immigrantion 2001-2011', y='Cases Received', data= immigrantion_2001_df, hue='Cases Received', palette='coolwarm', legend=False, ax=ax2)
ax2.set_title('Average Cases Received by Immigrantion between 2001 - 2011 of Total Population(%)')
ax2.set_xlabel('Immigrantion between 2001 - 2011(%)')
ax2.set_ylabel('Average Cases Received')

sns.barplot(x='Visible minority population', y='Cases Received', data= minority_df, hue='Cases Received', palette='coolwarm', legend=False, ax=ax3)
ax3.set_title('Average Cases Received by Minority Population of Total Population(%)')
ax3.set_xlabel('Minority Population(%)')
ax3.set_ylabel('Average Cases Received')
plt.tight_layout()
````
